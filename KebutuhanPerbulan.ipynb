```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

data = pd.read_csv('/content/kebutuhan_per_bulan_2.csv')

color_mapping = {'ACEH': 'purple',
    'SUMATERA UTARA': 'orange',
    'SUMATERA BARAT': 'lime',
    'RIAU': 'teal',
    'JAMBI': 'pink',
    'SUMATERA SELATAN': 'gold',
    'BENGKULU': 'blue',
    'LAMPUNG': 'red',
    'KEP. BANGKA BELITUNG': 'brown',
    'KEP. RIAU': 'gray',
    'DKI JAKARTA': 'magenta',
    'JAWA BARAT': 'cyan',
    'JAWA TENGAH': 'green',
    'DI YOGYAKARTA': 'navy',
    'JAWA TIMUR': 'olive',
    'BANTEN': 'maroon',
    'BALI': 'indigo',
    'NUSA TENGGARA BARAT': 'silver',
    'NUSA TENGGARA TIMUR': 'turquoise',
    'KALIMANTAN BARAT': 'chocolate',
    'KALIMANTAN TENGAH': 'coral',
    'KALIMANTAN SELATAN': 'khaki',
    'KALIMANTAN TIMUR': 'lavender',
    'KALIMANTAN UTARA': 'orchid',
    'SULAWESI UTARA': 'peru',
    'SULAWESI TENGAH': 'plum',
    'SULAWESI SELATAN': 'salmon',
    'SULAWESI TENGGARA': 'sienna',
    'GORONTALO': 'skyblue',
    'SULAWESI BARAT': 'tan',
    'MALUKU': 'thistle',
    'MALUKU UTARA': 'violet',
    'PAPUA': 'yellow',
    'PAPUA BARAT': 'darkgreen'
}

plt.figure(figsize=(20, 60))

# Mengurutkan data berdasarkan kolom 'kebutuhan'
sorted_data = data.sort_values('kebutuhan')

for i, row in sorted_data.iterrows():
    plt.scatter(row['tahun'], row['kebutuhan'], c=color_mapping[row['Propinsi']], s=100, alpha=0.5)
    plt.text(row['tahun'], row['kebutuhan'], f"{row['kebutuhan']}", fontsize=8, ha='center', va='bottom')

for kota, warna in color_mapping.items():
    plt.scatter([], [], c=warna, label=kota)

plt.legend()
plt.xlabel('Jenis')
plt.ylabel('Jumlah')
plt.title('Supply Kantor')

plt.show()
```
